## **ðŸ“Œ Step 1: Import Pandas & Create a Dataset**

```python
#import
import pandas as pd
# Creating Employee Data
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR'],
    'Salary': [50000, 70000, 65000, 72000, 48000],
    'Experience': [2, 5, 4, 6, 1]
}

# Convert to DataFrame
df=pd.DataFrame(data)

# Display DataFrame
df
```

## **ðŸ“Œ Step 2: Check Basic Info**

```python
# Check data types & missing values
df.info()
```

```python
# Summary statistics
df.describe()
```

## **ðŸ“Œ Step 3: Selecting & Filtering Data**

```python
# Select Salary column
df['Salary']
```

```python
# Filter employees earning > 60,000
df[df['Salary']>60000]
```

## **ðŸ“Œ Step 4: Sorting & Grouping**

```python
# Sort employees by Salary
df.sort_values(by='Salary',ascending=False)
```

```python
# Avg Salary per department
df.groupby('Department')['Salary'].mean()
```

## **ðŸ“Œ Step 5: Adding a New Column**

```python
# 10% hike
df["Hike"]=df['Salary']*1.1
df
```

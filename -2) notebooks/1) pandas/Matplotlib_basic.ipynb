# Matplotlib Basics - Jupyter Notebook

## 1. Introduction to Matplotlib
Matplotlib is a powerful Python library used for creating static, animated, and interactive visualizations.


## 2. Importing Matplotlib

```python

import matplotlib.pyplot as plt
import numpy as np
```

## 3. Basic Line Plot

```python


x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Basic Line Plot")
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Bar Chart Example

```python

categories = ["A", "B", "C", "D"]
values = [10, 15, 7, 12]

plt.bar(categories, values, color='green')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

## 5. Histogram Example

```python

data = np.random.randn(1000)  # Generating random data

plt.hist(data, bins=20, color='purple', alpha=0.7)
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

## 6. Customization in Matplotlib

```python

plt.figure(figsize=(8,5))  # Set figure size
plt.style.use('ggplot')    # Use a predefined style

plt.plot(x, y, color='r', marker='s', linestyle='--', label="Customized Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Plot")
plt.legend()
plt.grid(True)
plt.show()
```

## 7. Subplots Example

```python

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# First subplot - Line plot
axes[0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0].set_title("Line Plot")

# Second subplot - Bar chart
axes[1].bar(categories, values, color='orange')
axes[1].set_title("Bar Chart")

plt.tight_layout()
plt.show()
```

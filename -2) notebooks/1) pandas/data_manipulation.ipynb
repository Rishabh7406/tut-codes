```python

# HANDLING MISSING VALUES
# Concept: Missing data is common in datasets; handling it properly is crucial.
# Methods:
# •
# dropna(): Removes missing values (entire row/column if NaN is present)
# • fillna(): Fills missing values with a specified value (mean, median, mode, forward fill, backward fill)
# •
# ·
# isnull().sum(): Counts missing values in each column
# • interpolate(): Estimates missing values using interpolation
# techniques
# • replace(): Replaces missing or specific values with defined values
```

```python
import pandas as pd
import numpy as np
```

```python

url="https://raw.githubusercontent.com/Rishabh7406/tut-codes/refs/heads/main/-1)%20datasets/Expanded_data_with_more_features.csv"
df=pd.read_csv(url,index_col='Unnamed: 0')

```

```python
df.info()
```

```python
df.head()
```

```python
# Display missing values count
display(df.isnull().sum())
```

```python
### Handling Missing Values ###

# 1. dropna(): Remove rows with missing values
df_dropna = df.dropna()
display(df_dropna.head())
```

```python
# Remove columns with missing values
df_dropna_cols = df.dropna(axis=1)
display(df_dropna_cols.head())
```

```python
# 2. fillna(): Fill missing values with a specific value
# Fill numeric columns with mean
df_filled_mean = df.fillna(df.mean(numeric_only=True))
display(df_filled_mean.head())
```

```python

# Fill categorical columns with mode
df_filled_mode = df.fillna(df.mode().iloc[0])
display(df_filled_mode.head())
```

```python
# Forward fill (propagating last valid value forward)
df_ffill = df.fillna(method='ffill')
display(df_ffill.head())
```

```python
# Backward fill (propagating next valid value backward)
df_bfill = df.fillna(method='bfill')
display(df_bfill.head())
```

```python
# 3. isnull().sum(): Count missing values
display(df.isnull().sum())
```

```python

# 4. interpolate(): Estimate missing values
df_interpolated = df.interpolate()
display(df_interpolated.head())
```

```python


# 5. replace(): Replace specific values
df_replaced = df.replace({"NaN": "Unknown", None: "Missing"})
display(df_replaced.head())
```

```python
### Separate DataFrame for ffill, bfill, and interpolate Demonstration ###
df_demo = pd.DataFrame({
    "A": [1, np.nan, np.nan, 4, 5],
    "B": [10, 20, np.nan, np.nan, 50],
    "C": [np.nan, 200, 300, np.nan, 500]
})


```

```python

print("Original DataFrame:")
display(df_demo)
```

```python

# Forward Fill
df_ffill_demo = df_demo.fillna(method='ffill')
print("Forward Fill:")
display(df_ffill_demo)
```

```python

# Backward Fill
df_bfill_demo = df_demo.fillna(method='bfill')
print("Backward Fill:")
display(df_bfill_demo)
```

```python

# Interpolation
df_interpolate_demo = df_demo.interpolate()
print("Interpolation:")
display(df_interpolate_demo)
```

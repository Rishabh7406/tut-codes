# Matplotlib Practice Tasks

This notebook contains practice tasks for Matplotlib.
Complete each task step by step.

```python

import matplotlib.pyplot as plt
import numpy as np
```


# Task 1: Line Plot

```python

x = [1, 2, 3, 4, 5]
y = [5, 10, 5, 15, 10]
plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Task 1: Line Plot")
plt.grid(True)
plt.legend()
plt.show()
```


# Task 2: Bar Chart

```python

categories = ["A", "B", "C", "D"]
values = [12, 7, 15, 10]
plt.figure(figsize=(6,4))
plt.bar(categories, values, color='green')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Task 2: Bar Chart")
plt.show()
```


# Task 3: Histogram

```python

data = np.random.randn(500)  # Generating 500 random numbers
plt.figure(figsize=(6,4))
plt.hist(data, bins=20, color='purple', alpha=0.7)
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.title("Task 3: Histogram")
plt.show()
```


# Task 4: Scatter Plot

```python

x_scatter = np.random.rand(50) * 10  # Random x values
y_scatter = np.random.rand(50) * 10  # Random y values
plt.figure(figsize=(6,4))
plt.scatter(x_scatter, y_scatter, color='red', marker='x')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Task 4: Scatter Plot")
plt.show()
```

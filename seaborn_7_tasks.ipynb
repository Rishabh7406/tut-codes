# seaborn Practice Tasks

This notebook contains practice tasks for seaborn.
Complete each task step by step.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Load Titanic dataset

```python
# Load Titanic dataset
df = sns.load_dataset('titanic')
```


# Task 1: Passenger Class Distribution

```python

plt.figure(figsize=(6,4))
sns.countplot(x='pclass', data=df, palette='viridis')
plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

# Task 2: Average Fare by Embarkation Point (Bar Chart)

```python
plt.figure(figsize=(6,4))
embark_fares = df.groupby('embark_town')['fare'].mean().reset_index()
sns.barplot(x='embark_town', y='fare', data=embark_fares, palette='magma')
plt.title("Average Fare by Embarkation Point")
plt.xlabel("Embarkation Town")
plt.ylabel("Average Fare")
plt.show()
```

# Task 3: KDE Plot - Fare Distribution by Class

```python
plt.figure(figsize=(6,4))
sns.kdeplot(data=df, x='fare', hue='pclass', fill=True, common_norm=False, palette='Set1')
plt.title("Fare Distribution by Class")
plt.xlabel("Fare")
plt.show()
```

# Task 4: Find the Highest Fare Paid

```python
max_fare = df['fare'].max()
highest_payers = df[df['fare'] == max_fare]
print("Highest Fare Paid:", max_fare)
print(highest_payers[['survived', 'pclass', 'sex', 'fare', 'embark_town']])
```

# Task 5: Bar Plot of Top 10 Highest Fares

```python
top_fares = df[['fare', 'sex', 'pclass']].sort_values(by='fare', ascending=False).head(10)
plt.figure(figsize=(6,4))
sns.barplot(x='fare', y=top_fares.index, data=top_fares, hue='pclass', palette='rocket')
plt.title("Top 10 Highest Fares")
plt.xlabel("Fare")
plt.show()
```
